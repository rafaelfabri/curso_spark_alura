```python
produtos = spark.createDataFrame(
        [
            ('1', 'Bebidas', 'Agua'),
            ('2', 'Limpeza', 'Sabao'),
            ('3', 'Frios', 'Queijo'),
            ('4', 'Bebidas', 'Refri'),
            ('5', 'Pet', 'Racao'),
        ],
        ['id', 'cat', 'prod']
    )

impostos = spark.createDataFrame(
        [
            ('Bebidas', 0.15),
            ('Limpeza', 0.05),
            ('Frios', 0.06),
            ('Carnes', 0.08)
        ],
        ['cat', 'tax']
        
)
```

```python
produtos.show()
```

```python
impostos.show()
```

```python
produtos_join = produtos.join(impostos, on = 'cat', how = 'inner')

produtos_join.show()
```

```python
produtos_join = produtos.join(impostos, on = 'cat', how = 'left')

produtos_join.show()
```

```python
produtos.join(impostos, on = 'cat', how = 'right')\
    .sort('id')\
    .show()
```

```python
produtos.join(impostos, on = 'cat', how = 'outer')\
    .sort('id')\
    .show()
```

```python
estabelecimentos_join = estabelecimentos.join(empresas, on = 'cnpj_basico', how = 'inner')
```

```python
estabelecimentos_join.printSchema()
```

```python
freq = estabelecimentos_join\
    .select('cnpj_basico',
            f.year('data_de_inicio_atividade').alias('data_inicio')
           )\
    .where('data_inicio >= 2010')\
    .groupBy('data_inicio')\
    .agg(f.count('cnpj_basico').alias('frequencia'))\
    .orderBy('data_inicio', ascending = True)
```

```python
freq.toPandas()
```

```python
freq.union(
    freq.select(
        f.lit('Total').alias('data_incio'),
        f.sum(freq['frequencia']).alias('frequencia')
    )
).show()
```

```python
estabelecimentos_join.createOrReplaceTempView('empresasJoinView')
```

```python
freq = spark\
    .sql('''
    
    SELECT YEAR(data_de_inicio_atividade) AS DATA_INICIO, 
    COUNT(cnpj_basico) AS count
    from empresasJoinView
    where year(data_de_inicio_atividade) >= 2010
    GROUP BY DATA_INICIO
    ORDER BY DATA_INICIO
    ''')
```
